```python
filteredColumns = {
    columnName: columnValue
    for columnName, columnValue in columns.items()
    if "raw-sheet-name" in columnValue
}

rawSheetColumns = [name["raw-sheet-name"]["name"] for name in filteredColumns.values()]

pdfColumns = [
    element[0]
    for element in sorted(
        {
            columnName: columnValue
            for columnName, columnValue in columns.items()
            if "invoice-pdf" in columnValue and "index" in columnValue["invoice-pdf"]
        }.items(),
        key=lambda x: x[1]["invoice-pdf"]["index"],
    )
]

extractPDFColumns = [
    columnName
    for columnName, columnValue in columns.items()
    if "invoice-pdf" in columnValue and "raw-sheet-name" not in columnValue
]

df = df[rawSheetColumns]
df.columns = list(filteredColumns.keys())

# newDf["Sr"] = ""
for i in extractPDFColumns:
    df[i] = range(1, len(df) + 1)
```
