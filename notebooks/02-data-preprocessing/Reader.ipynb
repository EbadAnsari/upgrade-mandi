```python
from utils.reader import reader
import pandas as pd
```

```python
df = reader.read_excel("./utils/heavy.xlsx", "Sheet1")

df.columns = df.iloc[0]
df = df.drop(0, axis=0).reset_index(drop=True)
```

```python
def filter(row: pd.Series) -> bool:
    percentage_of_valid = 0
    for cell in row:
        if cell.strip() != "":
            percentage_of_valid += 1
    return percentage_of_valid / len(row) >= 0.7


filtered_df = df[df.apply(filter, axis=1)].reset_index(drop=True)
```
